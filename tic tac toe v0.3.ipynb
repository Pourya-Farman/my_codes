```python
def x_o():
    
    #take in player #1 input to select from X and O 

    user1_input_letter=input('Player 1, Please select: X or O? ')

    #verify the input

    while not user1_input_letter.capitalize()== 'X' and not user1_input_letter.capitalize()== 'O':
        
        user1_input_letter=input('Incorrect input Player 1, Please select: X or O? ') 
        
        user_input_player_one= user1_input_letter.capitalize()
 
    #assign X and O based on player 1 choice
    
    if user1_input_letter.capitalize()=='X':
        
        user_input_player_one='X'
        
        user_input_player_two='O'
    else:
        
        user_input_player_one='O'
        
        user_input_player_two='X'
    
    #print the final assignments 

    print(f'Player 1 is {user_input_player_one}')
    
    print(f'Player 2 is {user_input_player_two}')
    
    return [user_input_player_one, user_input_player_two]
```

```python
def tic_tac_toe():
    
    print("Let's play Tic Tac Toe")
    
    play_again1=True
    
    #this while loop is for checking if the user wants to repeat the game
    
    while play_again1:  
        
        x = 0

        row_column_existing_list = []

        #players select from X and O to take turns
        
        players_choice= x_o()
        
        #print the board for the first time
        
        lines=print_board()

        player_take_turn=True

        #this while loop is for checking if the game is tied or not
        
        while x < 9:

            if player_take_turn:
                player_turn= players_choice[0]

            else:
                player_turn= players_choice[1]

            # take input from user on row
            
            row = input(f'player {player_turn}, Please enter row: select a number (an integer) between 1 to 3')

            #validate the input 
            
            row_int = check_players_input_row(row)

            # take input from user on column
            
            column = input('Please enter column: select a number (an integer) between 1 to 3')

            #validate the input
            
            column_int = check_players_input_column(column)

            #check if the row and column have been selected before
            
            row_column = check_inp_vs_existing_list(row_int, column_int, row_column_existing_list)

            #update the board with the input and print it
            
            player_take_turn,lines=replace_input(row_column,lines,player_take_turn,player_turn)

            #check if the user won the game
            
            winner1=check_the_winner(player_turn,lines)

            if winner1:
                
                print(f'{player_turn} is the winner')
                break
            
            else:    
                
                x += 1

            if x==9:
                
                print('No winner, the game is tied')
        
        #check if the users want to play again
        
        play_again1=play_again()
```
